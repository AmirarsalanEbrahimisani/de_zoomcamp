```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 1. Data Loading
url = 'https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv'
df_raw = pd.read_csv(url)

# Select features
columns_to_use = [
    'engine_displacement',
    'horsepower',
    'vehicle_weight',
    'model_year',
    'fuel_efficiency_mpg'
]
data = df_raw[columns_to_use].copy()
display(data.head())
```

```python
# 2. Shuffling and Splitting (60/20/20)
np.random.seed(42)
idx = np.arange(len(data))
np.random.shuffle(idx)

# Reorder and reset index
data_shuffled = data.iloc[idx].reset_index(drop=True)

n = len(data_shuffled)
n_val = int(n * 0.2)
n_test = int(n * 0.2)
n_train = n - n_val - n_test

df_train = data_shuffled.iloc[:n_train].copy()
df_val = data_shuffled.iloc[n_train:n_train + n_val].copy()
df_test = data_shuffled.iloc[n_train + n_val:].copy()

print(f'Train: {len(df_train)}, Val: {len(df_val)}, Test: {len(df_test)}')
```

```python
# 3. Model Functions
def train_linear_regression(X, y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    XTX = X.T @ X
    w = np.linalg.inv(XTX) @ X.T @ y
    return w[0], w[1:]

def prepare_X(df, fillna_value):
    df_num = df.copy()
    df_num['horsepower'] = df_num['horsepower'].fillna(fillna_value)
    return df_num.values

def rmse(y, y_pred):
    return np.sqrt(((y - y_pred) ** 2).mean())
```

```python
# 4. Training and Evaluation (Imputing with Mean)
hp_mean = df_train['horsepower'].mean()
features = ['engine_displacement', 'horsepower', 'vehicle_weight', 'model_year']

y_train = df_train.fuel_efficiency_mpg.values
y_val = df_val.fuel_efficiency_mpg.values

# Prepare features
X_train = prepare_X(df_train[features], hp_mean)
w0, w = train_linear_regression(X_train, y_train)

# Validate
X_val = prepare_X(df_val[features], hp_mean)
y_pred = w0 + X_val @ w

print(f'Validation RMSE: {round(rmse(y_val, y_pred), 4)}')
```
