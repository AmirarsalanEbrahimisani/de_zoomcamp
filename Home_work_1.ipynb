Q7.Sum of weights

```python
columns = ['vehicle_weight', 'model_year']
X = cars_from_asia[columns].head(7).values # Matris işlemleri için .values (NumPy array) daha güvenlidir
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])

XT = X.T


XTX = XT @ X


XTX_inv = np.linalg.inv(XTX)


w = XTX_inv @ XT @ y

w.sum()
```
